```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
data=pd.read_csv("Ads.csv")
data
```

```python
data.index
```

```python
real_x=data.iloc[:,[2,3]].values # age and estimated salary
```

```python
real_y=data.iloc[:,4].values
```

```python
training_x,test_x,training_y,test_y=train_test_split(real_x,real_y,test_size=0.25,random_state=0)
training_x
```

```python
#features scaling"""
scalar=StandardScaler()
training_x=scalar.fit_transform(training_x)
test_x=scalar.fit_transform(test_x)
```

```python
training_x
```

```python
test_x
```

```python
classifier_lgreg=LogisticRegression(random_state=0)
classifier_lgreg.fit(training_x,training_y)
```

```python
y_pred=classifier_lgreg.predict(test_x)
```

```python
y_pred
```

```python
test_y
```

```python
cnm=confusion_matrix(test_y,y_pred)
```

```python
cnm
test_y.size
```

```python
test_x.size
```

```python
#from matplotlib.colors import ListedColormap

x_set,y_set=training_x,training_y
X1,X2=np.meshgrid(np.arange(start=x_set[:,0].min()-1,stop=x_set[:,0].max()+1,step=0.01),
                  np.arange(start=x_set[:,1].min()-1,stop=x_set[:,1].max()+1,step=0.01))
plt.contourf(X1,X2,classifier_lgreg.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),
            alpha=0.75,camp=ListedColormap(('red','green')))
plt.xlim(X1.min(),X1.max())
plt.ylim(X2.min(),X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set==j,0],x_set[y_set==j,1],
               c=ListedColormap(('red','green'))(i),label=j)
plt.title('Logistic Regression(Training Set)')
plt.xlabel('age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()

                  
```
